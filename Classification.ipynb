```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.decomposition import PCA
from sklearn import svm
```

```python
train = np.load('data/class_train_in.npy')
label = np.load('data/class_train_out.npy')
dp = np.delete(train, 0, 1)
lbl = np.delete(label,0,1)
lbl = [x[0] for x in lbl.tolist()]
#tr = dp[]
#subdata
X = dp[0:100,0:2]
y = lbl[:100]
```

PCA

```python
pca = PCA(n_components=2)
X_r = pca.fit(dp).transform(dp)
```

```python
for p,l in zip(X_r, lbl):
    if l[0] == 0:
        plt.scatter(p[0],p[1], color = 'r')
    else:
        plt.scatter(p[0],p[1], color = 'b')
plt.show()
```

Linear SVM

```python
C = 1
svc = svm.SVC(kernel='linear', C=C).fit(X, y)
```

```python
h = .2  # step size in the mesh
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min, y_max, h))
```
